```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Загрузка данных из CSV файла
df = pd.read_csv('NPHA-doctor-visits.csv')

print(f'Статистическая информация:\n{df.describe()}\n')

print(f'Названия столбцов:\n{df.columns}\n')

# Дополнительная информация о датафрейме
print(f'{df.info()}\n')

"""Все данные типа int, нулевые значения отсутсвуют --> преобразования не требуются"""
df = df[df['Dental Health'] != 6] # В описании df отсутствует значение 6, поэтому, удаляем
df = df[df['Trouble Sleeping'] != -1] # аналогично, удалил неизвестные значения из другого признака
df = df[df['Trouble Sleeping'] != 3]

# Визуализация распределения возраста бессмыслена - все значения одинаковые

# Визуализация количества визитов к врачу
plt.figure(figsize=(12, 6))
sns.countplot(x='Number of Doctors Visited', data=df)
plt.title('Кол-во врачей, которых посетил пациент(1 - 0 или 1, 2 - 2 или 3, 3 - 4+)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация оценки физического здоровья
plt.figure(figsize=(12, 6))
sns.countplot(x='Phyiscal Health', data=df)
plt.title('Физическое здоровье (-1: Отказано 1: Отлично 2: Очень хорошо 3: Хорошо 4: Справедливо 5: Плохо)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация оценки ментального здоровья
plt.figure(figsize=(12, 6))
sns.countplot(x='Mental Health', data=df)
plt.title('Ментальное здоровье (-1: Отказано 1: Отлично 2: Очень хорошо 3: Хорошо 4: Справедливо 5: Плохо)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация самостоятельной оценки полости рта или зубов
plt.figure(figsize=(12, 6))
sns.countplot(x='Dental Health', data=df)
plt.title('Самостоятельная оценка зубов и полости рта(-1: Отказано 1: Отлично 2: Очень хорошо 3: Хорошо 4: Справедливо 5: Плохо)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация занятости населения
plt.figure(figsize=(12, 6))
sns.countplot(x='Employment', data=df)
plt.title('Статус занятости(-1: Отказано 1: Работа полный рабочий день 2: Работа на неполный рабочий день 3: Завершил карьеру 4. В настоящее время не работает)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация влияния стресса на сон
plt.figure(figsize=(12, 6))
sns.countplot(x='Stress Keeps Patient from Sleeping', data=df)
plt.title('Влияет ли стресс на способность пациента спать?(0: нет 1: да)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация влияния лекарств на сон
plt.figure(figsize=(12, 6))
sns.countplot(x='Medication Keeps Patient from Sleeping', data=df)
plt.title('Влияет ли лекарство на сон пациента?(0: нет 1: да)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация влияния физической боли на сон
plt.figure(figsize=(12, 6))
sns.countplot(x='Pain Keeps Patient from Sleeping', data=df)
plt.title('Нарушает ли физическая боль сон пациента?(0: нет 1: да)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация влияния физической боли на сон
plt.figure(figsize=(12, 6))
sns.countplot(x='Pain Keeps Patient from Sleeping', data=df)
plt.title('Нарушает ли физическая боль сон пациента?(0: нет 1: да)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация влияния необходимости посещения туалета на сон
plt.figure(figsize=(12, 6))
sns.countplot(x='Bathroom Needs Keeps Patient from Sleeping', data=df)
plt.title('Влияет ли необходимость посещения туалета на сон пациента?(0: нет 1: да)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация влияния неизвестных причин на сон
plt.figure(figsize=(12, 6))
sns.countplot(x='Uknown Keeps Patient from Sleeping', data=df)
plt.title('Влияют ли неизвестные причины на сон пациента?(0: нет 1: да)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация наличия проблем со сном
plt.figure(figsize=(12, 6))
sns.countplot(x='Trouble Sleeping', data=df)
plt.title('Есть ли проблемы со сном у пациента?(1: нет 2: да)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация применения препаратов для сна
plt.figure(figsize=(12, 6))
sns.countplot(x='Prescription Sleep Medication', data=df)
plt.title('Информация о назначенных препаратах(1: Отказано 1. Используйте регулярно 2. Используйте время от времени 3: Не используйте')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация по расе
plt.figure(figsize=(12, 6))
sns.countplot(x='Race', data=df)
plt.title('Раса (-2: Не спрашивается -1: ОТКАЗАЛСЯ 1: Белый, неиспаноязычный 2: Черный, неиспаноязычный 3: Другие, неиспаноязычные 4: Латиноамериканец 5: 2+ расы, неиспаноязычные)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Визуализация по полу
plt.figure(figsize=(12, 6))
sns.countplot(x='Gender', data=df)
plt.title('Пол пациента(-2: Не спрашивается -1: ОТКАЗАЛСЯ 1: Мужской 2: Женский)')
plt.xlabel('Оценка')
plt.ylabel('Кол-во пациентов')
plt.show()

# Групповая визуализация по расе и ментальному здоровью
plt.figure(figsize=(12, 6))
sns.barplot(x='Race', y='Mental Health', data=df)
plt.title('Психическое здоровье по расе')
plt.show()

# Групповая визуализация по наличию проблем со сном и физическим здоровьем
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Trouble Sleeping', y='Phyiscal Health');
plt.title('Влияние проблем со сном на физическое здоровье')
plt.show()

# Тепловая карта корреляции признаков
heatmap = df.corr()

"""Вывод - признаки слабо коррелируют между собой"""
```
